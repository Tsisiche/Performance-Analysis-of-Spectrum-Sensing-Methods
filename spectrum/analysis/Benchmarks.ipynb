```python
from matplotlib import cm
import re

Ns = [25000, 50000, 100000]

cmap = cm.get_cmap('gist_rainbow')

time = []
n = 2

for Ns1 in Ns:
    time1 = []
    methods = []
    left = []
    color = []
    left1 = 0
    i = 0
    j = 0
    for line in open("../measurements/benchmark/benchmark-laptop.dat"):
        f = line.split()
        
        Ns0 = int(f[1])
        if Ns0 == Ns1:
            m = f[0]
            m = m.replace('(', ' ').replace(')', '')
            m = m.replace('EnergyDetector', 'ED')
            m = m.replace('Detector', '')
            m = re.sub("L=(\d+)", r"$L=\1$", m)
            
            
            c = cmap(i/7. + j*0.007)
            #color.append(c)
            color.append(.85)
            
            methods.append(m)
            
            time1.append(float(f[2]))
            
            left.append(left1)
            left1 += 1
            j += 1
            
            if n % 4 == 0:
                left1 += 1
                i += 1
                j = 0
            n += 1
            
    time.append(time1)
    
left = array(left)
time = array(time)
```

```python
def plot_time(i):
    height = time[i,:]/time[i,0]
    
    figure(figsize=(13, 5))

    bar(left, height, color=color)
    xticks(left+.4, methods, rotation=60, horizontalalignment='right')
    xlim(-.5, left1-.5)
    ylim(0, 14)
    
    for x, y in zip(left, height):
        text(x+.4, y+1., "%.1f" % y, horizontalalignment="center", rotation=70, backgroundcolor='w')
    
    ylabel("$t_{CPU}$ [relative to energy detector]")
    title("$t=25.0$ms ($N_s = %d$ksamples)" % (Ns[i]/1e3,))
    
    tight_layout()
    grid()
    
plot_time(0)
#savefig("time_comparison_25ms.png", dpi=120)
#savefig("time_comparison_25ms.eps")
```

```python
plot_time(1)
#savefig("time_comparison_13ms.png", dpi=120)
```
